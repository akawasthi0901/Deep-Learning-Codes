```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sn
```

```python
import warnings
warnings.filterwarnings('ignore')

df=pd.read_csv("dltuts10regularization_csvsheet.csv",header=None)
df
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
#60th column is the target column we are checking the different values in the target 
df[60].value_counts()
```

```python
X=df.drop(60,axis='columns')
y=df[60]
y.head()
```

```python
X.head()
```

```python
#onehotencoding for y it has R and M
```

```python
y=pd.get_dummies(y,drop_first=True)
y.sample(5) #R--1,M--0
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
X_train.shape,X_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model=keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])

model.compile(
loss='binary_crossentropy',
optimizer='adam',
metrics=['accuracy'])

model.fit(X_train,y_train,epochs=100,batch_size=8)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred= model.predict(X_test).reshape(-1)
print(y_pred[:10])

#round the values to nearest integer 0 or 1
y_pred=np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix,classification_report

print(classification_report(y_test,y_pred))
```

```python
modeld=keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1,activation='sigmoid')
])

modeld.compile(
loss='binary_crossentropy',
optimizer='adam',
metrics=['accuracy'])

modeld.fit(X_train,y_train,epochs=100,batch_size=8)
```

```python
modeld.evaluate(X_test,y_test)
```

```python
y_pred = modeld.predict(X_test).reshape(-1)
y_pred = np.round(y_pred)
from sklearn.metrics import confusion_matrix,classification_report

print(classification_report(y_test,y_pred))
```
